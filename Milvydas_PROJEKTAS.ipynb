# Klausimas nr. 3
Girtų/Apsvaigusių vairuotojų kiekį. Ar girti vairuotojai sukeldavo daugiau eismo įvykių, kur būdavo sužeidžiami žmonės?

```python
appended['neblaivusKaltininkai'].unique()
```

```python
def girtiems(eilute):
  if eilute['neblaivusKaltininkai']=='Ne':
    return 0
  elif eilute['neblaivusKaltininkai']=="Taip":
    return 1

appended['girtas'] = appended.apply(girtiems, axis=1)
appended['girtas']
```

```python
appended['apsvaigeKaltininkai'].unique()
```

```python

def apsvaigus(eilute):
  if eilute['apsvaigeKaltininkai']=='Ne':
    return 0
  elif eilute['apsvaigeKaltininkai']=="Taip":
    return 1

appended['apsvaige'] = appended.apply(apsvaigus, axis=1)
appended['apsvaige']
```

```python
apsvaiges=appended['apsvaige'].sum()
girti=appended['girtas'].sum()
print(f"Viso girtų vairuotojų [2016-2021] metais: {girti}.")
print(f"Viso apsvaigusių vairuotojų [2016-2021] metais: {apsvaiges}.")
print(f"Viso girtų ar apsvaigusių vairuotojų [2016-2021] metais: {apsvaiges + girti}.")
```

# Galutinės išvados

2016 - 2021 metais įvyko 142907 eismo įvykiai, tarp kurių 17936 įskaitiniai(nukentėjo žmonės). Nors eismo įvykių padaugėję(lyginant 2016 ir 2021 metus), tačiau įskaitinių(kai nukenčia žmonės) eismo įvykių mažėja. Dažniausia rūšis – Susidūrimas. Nuo jos neatslieka: susidūrimai su stovinčia transporto priemone bei užvažiavimas ant gyvūno. 2016 - 2021 metais beveik kas devintame eismo įvykyje buvo nukentėjęs žmogus.

2016 - 2021 metais žuvusių skaičius krito, o sužeistų krito iki 2020, tačiau vėl pakilo 2021. Deja, bet žuvusių vaikių skaičius išlieka panašus, o sužeistų vaikų dalis lyginant su visais sužeistaisiais 2021 metais pati didžiausia. 2016 – 2021 beveik kas 11 eismo įvykio dalyvis nukenčia. Iš visų nukentėjusių net ~16 procentų - vaikai.

2016 – 2021 metais iš kaltų dėl eismo įvykio vairuotų beveik kas 16-tas buvo apsvaigęs arba neblaivus. Galime pasidžiaugti, kad girtų ar apsvaigusių vairuotojų skaičius mažėja ir sužeistų skaičius mažėja, tačiau dėl jų kaltės žūsta vis daugiau žmonių. Beveik kas 9 žuvęs žmogus ir beik kas 13 sužeistas žmogus yra dėl girto vairuotojo kaltės.
2016 – 2021 eismo įvykių skaičius išlieka panašus esant tam tikrai oro sąlygai, žmonės nepasimoko iš klaidų, kaip elgtis esant tam tikram orui. Truputį mažiau nei pusė eismo įvykių įvyksta esant prastoms oro sąlygoms, kas leidžia teigti, kad oro sąlygos turi didelę įtaką iesmo įvykiams.
